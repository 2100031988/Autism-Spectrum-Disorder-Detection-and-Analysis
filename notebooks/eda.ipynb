```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('../data/autism_data.csv')

# Display the first few rows
df.head()

# Check for missing values
print("Missing values per column:")
print(df.isnull().sum())

# Select only numeric columns for correlation matrix (ignore object/categorical columns)
df_numeric = df.select_dtypes(include=['float64', 'int64'])

# Plot correlation matrix (for numeric data only)
plt.figure(figsize=(10, 8))
sns.heatmap(df_numeric.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()

# Visualize feature distributions for numeric columns
df_numeric.hist(figsize=(10, 10))
plt.tight_layout()
plt.show()
```
